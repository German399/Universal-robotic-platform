```python
from PIL import Image as pil
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/media/giordanolaminetti/dataset/TUM_SOLO/val_selection_cropped/image/1341848023.954826.png")

# You may need to convert the color.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
slam = np.array(pil.open("/media/giordanolaminetti/dataset/TUM_SOLO/val_selection_cropped/slam/1341848023.954826.png"))
print((slam>0).sum())
```

```python
index = np.argwhere(slam>0)
for point in index:
    img=cv2.circle(img,(int(point[1]),int(point[0])),10,(255,0,0),3)
plt.imshow(img)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
cv2.imwrite('out.png',img)
```
